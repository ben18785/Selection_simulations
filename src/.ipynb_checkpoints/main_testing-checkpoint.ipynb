```python
aNumGenerations = 100
aPopulation = Bentley.createInitialPopulation(10, 10)
a = aPopulation.getAllFreq()
lAdded = np.zeros(aNumGenerations, dtype='int')
lKilled = np.zeros(aNumGenerations, dtype='int')
print(aPopulation.getAllFreq())
for generations in range(aNumGenerations):
    aPopSize = aPopulation.getPopSize()
    weighted_choices = aPopulation.getAllFreq()
    p = 1 + (np.array(weighted_choices.values()) /
                 float(sum(weighted_choices.values())) * 0)
    p = p / sum(p)
    aParent_number = np.random.choice(weighted_choices.keys(), size=1, p=p)[0]
    aIndividual = Bentley.createOffspring_simple(aParent_number, 0.01,
                                                 aPopulation)
    aPopulation.addIndividual(aIndividual)
    lParents = aPopulation.getIndividuals()
    lAdded[generations] = aIndividual.getNumber()
    bRand = np.random.randint(aPopSize)
    aPopulation.killIndividual(lParents[bRand])
    lKilled[generations] = lParents[bRand].getNumber()
    aPopulation.calculateAllFreq()
print(aPopulation.getAllFreq())
```

```python
c = collections.Counter(lAdded)
d = collections.Counter(lKilled)
a = a + c
a.subtract(d)

print(c)
print(d)
print(a)
```
